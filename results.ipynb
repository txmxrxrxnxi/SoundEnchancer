Імпортування бібліотек

```python
import os
import pandas as pd
import matplotlib.pyplot as plt
```

Завантаження даних

```python
df = pd.read_csv("comparison_results.csv")
df
```

```python
df.describe()
```

```python
plt.figure(figsize=(10, 6))
for column in df.columns[1:]:
    plt.plot(df["File"], df[column], marker="o", label=column)
plt.xlabel("Файл")
plt.ylabel("Значення")
plt.title("Порівняння із зашумленими звуковими доріжками")
plt.xticks([])
plt.legend()
plt.show()
```

```python
df_clear = pd.read_csv("comparison_results.csv")
df_clear
```

```python
df_clear.describe()
```

```python
plt.figure(figsize=(10, 6))
df_clear["File"] = df_clear["File"].apply(lambda x: os.path.splitext(x)[0])
for column in df_clear.columns[1:]:
    plt.plot(df_clear["File"], df_clear[column], marker="o", label=column)
plt.xlabel("Файл")
plt.ylabel("Значення")
plt.title("Порівняння з чистими звуковими доріжками")
plt.xticks([])
plt.legend()
plt.show()
```
