Імпортування бібліотек

```python
import os
import pandas as pd
import matplotlib.pyplot as plt
```

Завантаження даних

```python
df = pd.read_csv("comparison_results.csv")
df
```

```python
df.describe()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(df["Файл"], df["Різниця центроїди з фільтром Вінера"], marker="o", label="Фільтр Вінера")
plt.plot(df["Файл"], df["Різниця цетнроїда з бібліотечним фільтром"], marker="o", label="Бібліотечний фільтр Вінера")
plt.xlabel("Файл")
plt.ylabel("Значення")
plt.title("Порівняння із зашумленими звуковими доріжками, спектральна центроїда (%)")
plt.xticks([])
plt.legend()
plt.show()

plt.figure(figsize=(10, 6))
plt.plot(df["Файл"], df["Різниця середньої площинності з фільтром Вінера"], marker="o", label="Фільтр Вінера")
plt.plot(df["Файл"], df["Різниця середньої площинності з бібліотечним фільтром"], marker="o", label="Бібліотечний фільтр Вінера")
plt.xlabel("Файл")
plt.ylabel("Значення")
plt.title("Порівняння із зашумленими звуковими доріжками, середня спектральна площинність (%)")
plt.xticks([])
plt.legend()
plt.show()
```

```python
df_absolute = pd.read_csv("absolute_results.csv")
df_absolute
```

```python
df_absolute.describe()
```

```python
plt.figure(figsize=(10, 6))

plt.plot(df_absolute["Файл"], df_absolute["Центроїд чистої"], marker="o", label="Чистої доріжки")
plt.plot(df_absolute["Файл"], df_absolute["Центроїд фільтра Вінера"], marker="o", label="Фільтра Вінера")
plt.plot(df_absolute["Файл"], df_absolute["Центроїд бібліотечного фільтра"], marker="o", label="Бібліотечного фільтра Вінера")
plt.plot(df_absolute["Файл"], df_absolute["Центроїд зашумленої"], marker="o", label="Зашумленої доріжки")
plt.xlabel("Файл")
plt.ylabel("Значення")
plt.title("Абсолютні значення спектральних центроїд")
plt.xticks([])
plt.legend()
plt.show()

plt.figure(figsize=(10, 6))
plt.plot(df_absolute["Файл"], df_absolute["Середня площинність чистої"], marker="o", label="Чистої доріжки")
plt.plot(df_absolute["Файл"], df_absolute["Середня площинність фільтра Вінера"], marker="o", label="Фільтра Вінера")
plt.plot(df_absolute["Файл"], df_absolute["Середня площинність бібліотечнрго фільтра"], marker="o", label="Бібліотечного фільтра Вінера")
plt.plot(df_absolute["Файл"], df_absolute["Середня площинність зашумленої"], marker="o", label="Зашумленої доріжки")
plt.xlabel("Файл")
plt.ylabel("Значення")
plt.title("Абсолютні значення середніх спектральних площинностей")
plt.xticks([])
plt.legend()
plt.show()
```
